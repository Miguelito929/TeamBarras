```python
import numpy as np
```

```python
def Distancia_Euclidiana(punto1, punto2):
    x0,y0 = punto1
    x1,y1 = punto2
    return ((x0-x1)**2 + (y0-y1)**2)**(1/2)

def Distancia_Manhattan(punto1, punto2):
    x0,y0 = punto1
    x1,y1 = punto2
    return (abs(x0-x1) + abs(y0-y1))  
```

```python
def inner_point_euclides(Datos, n, centro, radio):
    dentro = []
    indices = []    
    for i in range(n):
        if Distancia_Euclidiana(Datos[i],centro) < radio:
            dentro.append(Datos[i])
            indices.append(i)    
    return np.array(dentro)

def inner_point_manhattan(Datos, n, centro, radio):
    dentro = []
    indices = []    
    for i in range(n):
        if Distancia_Manhattan(Datos[i],centro) < radio:
            dentro.append(Datos[i])
            indices.append(i)   
    return np.array(dentro)
```

## Distancia Euclidiana

```python
n = 100 # NÃºmero de Datos
Datos = np.random.rand(n,2) * 10

bola1 = inner_point_euclides(Datos, n = 100, centro = [2,3], radio = 2.5)
bola2 = inner_point_euclides(Datos, n = 100, centro = [5,7], radio = 2.5)
bola3 = inner_point_euclides(Datos, n = 100, centro = [8,2], radio = 3)
inner_points = list(bola1) + list(bola2) + list(bola3)
aislados = np.array([list(x) for x in Datos if x not in np.array(inner_points)])
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Arial'
plt.style.use('ggplot')  # GGplot la mera cabra
colores = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3']
plt.figure(figsize=(10, 10))
plt.figure(figsize = (10,10))
plt.scatter(bola1[:,0], bola1[:,1], color = colores[0],label = 'Inner Points: Bola 1')
plt.scatter(bola2[:,0], bola2[:,1], color = colores[1],label = 'Inner Points: Bola 2')
plt.scatter(bola3[:,0], bola3[:,1], color = colores[2],label = 'Inner Points: Bola 3')
plt.scatter(aislados[:,0], aislados[:,1], color = colores[3],label = 'Puntos Aislados')    
centros = np.array([[2,3], [5,7], [8,2]])
radio = np.array([2.5, 2.5, 3])

i = 0
for centros, radius in zip(centros, radio):
    circle = plt.Circle(centros, radius, color = colores[i], fill = False, linestyle = '--')
    plt.gca().add_patch(circle)
    plt.annotate(f'Bola {i + 1}', xy=(centros[0], centros[1] + radius + 0.2), ha='center')
    i += 1
        
# TÃ­tulo
plt.legend() # loc='upper right

plt.suptitle(
    'Puntos interiores en Bolas Abiertas',
    color='black',
    fontsize=15,
    #loc='center',
    fontweight='bold',
    style='normal',
    family='Arial',
    y=0.950)

nota = '''Nota: Puntos interiores con el criterio de distancia Euclidiana'''
plt.figtext(0.5, 0.9, nota, ha='center', fontsize=12, style ='italic', family='Arial')
plt.show()
```

## Distancia Manhattan

```python
def edge_manhattan(Datos, n, centro, radio):
    dentro = []
    indices = []    
    for i in range(n):
        if abs(Distancia_Manhattan(Datos[i],centro)-radio) < .05:
            dentro.append(Datos[i])
            indices.append(i)
    return np.array(dentro)
```
